```python
train_path = 'drive/MyDrive/CODE/ViMMRC/dataset/original_data (for sharing)/train.json'
dev_path = 'drive/MyDrive/CODE/ViMMRC/dataset/original_data (for sharing)/dev.json'
test_path = 'drive/MyDrive/CODE/ViMMRC/dataset/original_data (for sharing)/test.json'
```

```python
import json

with open(train_path, 'r') as f:
    TRAIN = json.load(f, encoding='utf8')

with open(dev_path, 'r') as f:
    DEV = json.load(f, encoding='utf8')

with open(test_path, 'r') as f:
    TEST = json.load(f, encoding='utf8')
```

# Full

```python
reasons = ['Matching', 'Paraphrasing', 'Single-sentence Reasoning', 'Multi-sentence Reasoning', 'Ambiguous Or Insufficient']
choices = ['A', 'B', 'C', 'D']

def convert_race(DATA):
    race = []
    for d in DATA:
        ans = []
        ques = []
        types = []
        opts = []

        for q in d['questions']:
            if (q['question'].strip()):
                ops = []
                ans.append(q['answer'])
                ques.append(q['question'])
                for i in range(0, len(q['types'])):
                    if q['types'][i][reasons[i]] == True:
                        types.append(reasons[i])

                for o in q['options']:
                    oo = []
                    for c in choices:
                        if o.get(c) != None:
                            ops.append(o.get(c))
                opts.append(ops)
        race.append({
            'answers': ans,
            'options': opts,
            'questions': ques,
            'article': d['content'],
            'grade': d['grade'],
            'author': d['author'],
            'files': d['files'],
            'isProse': d['isProse'],
            'title': d['title'],
            'types': types
        })
    return race
```

```python
TRAIN_RACE = convert_race(TRAIN)
DEV_RACE = convert_race(DEV)
TEST_RACE = convert_race(TEST)
```

# Poem

```python
reasons = ['Matching', 'Paraphrasing', 'Single-sentence Reasoning', 'Multi-sentence Reasoning', 'Ambiguous Or Insufficient']
choices = ['A', 'B', 'C', 'D']

def convert_race_poem(DATA):
    race = []
    for d in DATA:
        if not d['isProse']:
            continue
        ans = []
        ques = []
        types = []
        opts = []

        for q in d['questions']:
            if (q['question'].strip()):
                ops = []
                ans.append(q['answer'])
                ques.append(q['question'])
                for i in range(0, len(q['types'])):
                    if q['types'][i][reasons[i]] == True:
                        types.append(reasons[i])

                for o in q['options']:
                    oo = []
                    for c in choices:
                        if o.get(c) != None:
                            ops.append(o.get(c))
                opts.append(ops)
        race.append({
            'answers': ans,
            'options': opts,
            'questions': ques,
            'article': d['content'],
            'grade': d['grade'],
            'author': d['author'],
            'files': d['files'],
            'isProse': d['isProse'],
            'title': d['title'],
            'types': types
        })
    return race
```

```python
TRAIN_RACE_POEM = convert_race_poem(TRAIN)
DEV_RACE_POEM = convert_race_poem(DEV)
TEST_RACE_POEM = convert_race_poem(TEST)
```
